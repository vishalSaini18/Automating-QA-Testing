```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random
import numpy as np
```

```python

def filter12(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[6]/div[2]/div[1]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    driver.execute_script("arguments[0].click();", checkbox)
    
def filter13(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[6]/div[2]/div[2]/div/button"))
    )
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
 
def get_low(driver):
    low = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[6]/div[2]/div[2]/div[2]/div/div[2]/ul/li[1]/div/span[2]"))
    )

    # Get the text from the element
    text_value = low.text

    # Print the first character (index 0)
    return text_value[0]

def get_middle(driver):
    low = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[6]/div[2]/div[2]/div[2]/div/div[2]/ul/li[2]/div/span[2]"))
    )

    # Get the text from the element
    text_value = low.text

    # Print the first character (index 0)
    return text_value[0]  # returns first two characters (indices 0 and 1)

def get_high(driver):
    low = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[6]/div[2]/div[2]/div[2]/div/div[2]/ul/li[3]/div/span[2]"))
    )

    # Get the text from the element
    text_value = low.text

    # Print the first character (index 0)
    return text_value[0]
```
