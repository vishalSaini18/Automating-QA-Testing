```python
# Filter-1(Location)
def filter1(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "//div[@class='space-y-2']//div[1]//div[2]//div[1]//div[1]//button[1]"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
 
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time

    # Wait for the plus icon to be clickable again after filtering
    try:
        plus_icon_after = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@class='space-y-2']//div[1]//div[2]//div[1]//div[1]//button[1]"))
        )
        driver.execute_script("arguments[0].click();", plus_icon_after)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter2(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[2]/div[2]/div[1]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[2]/div[2]/div[1]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter3(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html[1]/body[1]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/div[2]/div[2]/div[1]/button[1]"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html[1]/body[1]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/div[1]/div[2]/div[2]/div[2]/div[1]/button[1]"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter4(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[2]/div[2]/div[3]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[2]/div[2]/div[3]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")


    return elapsed_time
 
def filter5(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[3]/div[2]/div[1]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[3]/div[2]/div[1]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter6(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[3]/div[2]/div[2]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[3]/div[2]/div[2]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter7(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[3]/div[2]/div[3]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[3]/div[2]/div[3]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter8(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[4]/div[2]/div/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
 
    # Wait for the train to be loaded
    trians = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//span/span"))
    )
    no_of_trains = len(trians)
    i = random.randint(1, no_of_trains)
 
    trian_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li{[i]}//div//span/span"))
    )
   
    # Click the trian icon
    driver.execute_script("arguments[0].click();", trian_icon)
 
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[4]/div[2]/div/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter9(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[5]/div[2]/div[1]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[5]/div[2]/div[1]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter10(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[5]/div[2]/div[2]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[5]/div[2]/div[2]/div/button"))
        )
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
        # print("done")
    except:
        print("not closed")

    return elapsed_time
 
def filter11(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[5]/div[2]/div[3]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
 
    start_time = time.time()
 
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[5]/div[2]/div[3]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter12(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[6]/div[2]/div[1]/div/button"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
    start_time = time.time()
 
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    
    try:    # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/div[2]/div/div/div/div/div/div[1]/div/div[2]/div[2]/div/div/div[6]/div[2]/div[1]/div/button"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
 
def filter14(driver):
    # Wait for the plus icon to be clickable
    plus_icon = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, "/html[1]/body[1]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/div[1]/div[7]/div[2]/div[1]/div[1]/button[1]"))
    )
   
    # Click the plus icon
    driver.execute_script("arguments[0].click();", plus_icon)
    # Wait for the checkboxes to be loaded
    checkboxes = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.XPATH, "//ul//li//div//input"))
    )
    no_of_checkboxes = len(checkboxes)
    i = random.randint(1, no_of_checkboxes)
    # Wait for the checkbox to be clickable
    checkbox = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, f"//ul//li[{i}]//div//input"))
    )
    start_time = time.time()
 
    driver.execute_script("arguments[0].click();", checkbox)
    # IMPORTANT: Wait explicitly for filtering to finish
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[contains(@class,'text-white')]"))
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    try:
        # Wait for the plus icon to be clickable
        plus_icon = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.XPATH, "/html[1]/body[1]/div[2]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/div[1]/div[7]/div[2]/div[1]/div[1]/button[1]"))
        )
    
        # Click the plus icon
        driver.execute_script("arguments[0].click();", plus_icon)
    except Exception as e:
        print(f"Exception occurred while clicking plus_icon after filtering: {e}")

    return elapsed_time
```
